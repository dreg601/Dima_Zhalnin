```python
import pandas as pd
import io
import matplotlib.pyplot as plt
import numpy as np 
%matplotlib inline
import seaborn as sns
import scipy as sp
```

Создаю датафреймы и объединяю их.

```python
fuel= pd.read_excel('тест.xlsx', sheet_name='fuel')
types= pd.read_excel('тест.xlsx', sheet_name='fuel_types')
fuel_types = fuel.merge(types, left_on = 'FL_PROD_COD', right_on = 'prod_code')
```

```python
fuel_types
```

Задание 1: Вывести среднесуточный пролив по каждой АЗС.

```python
fuel_types.FL_EVEDATE.unique()
```

В данном случае, у нас одна дата. Значит, считаем среднее по всему датафрейму с группировкой только по АЗС.

```python
fuel_types.groupby(['FL_SMAZS']).FL_VOLUME.mean().reset_index()
```

Задание 2: Вывести объем топлива в разрезе типа топлива отсортированый от большего к меньшему

```python
fuel_types.groupby(['prode_name']).FL_VOLUME.sum().reset_index().sort_values('FL_VOLUME', ascending=False)
```

Задание 3: Вывести процент каждого вида топлива от общего числа по каждой АЗС

```python
fuel_types['FL_PERCENT'] = fuel_types['FL_VOLUME'] / fuel_types.groupby(['FL_SMAZS']).FL_VOLUME.transform('sum')
fuel_types.groupby(['FL_SMAZS', 'prode_name']).FL_PERCENT.sum().reset_index()
```

Задание 4: Вычеслить кол-во операций по каждому часу и  коду оплаты

```python
fuel_types['HOUR'] = fuel_types['FL_EVETIME'].str.slice(stop=2)
```

```python
fuel_types.groupby('HOUR').FL_PAY.count().reset_index()
```
